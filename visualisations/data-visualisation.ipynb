```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_shopping_market = pd.read_csv('../data/cleaned_dataset.csv', delimiter=';')
```

```python
df_shopping_market
```

```python
fig, ax = plt.subplots(figsize=(38, 20))
ax.bar(df_shopping_market['name'], df_shopping_market['daily_visitors'], color='k')
# Adjust tick parameters for the x-axis
ax.tick_params(axis='x', width=2, labelsize=25, labelrotation=45)
# Move x-tick labels to the left
for label in ax.get_xticklabels():
    label.set_ha('right')  # Align the label to the right
    label.set_position((label.get_position()[0] - 0.02, label.get_position()[1]))
ax.set_ylabel('daily_visitors', weight='bold', fontsize=30, labelpad=30)
ax.tick_params(axis='y', width=2, labelsize=25)
ax1 = ax.twinx()
ax1.plot(df_shopping_market['name'], df_shopping_market['openingYear'], color='#b60000', marker='D')
ax1.set_ylabel('openingYear', weight='bold', fontsize=30, color='#b60000', labelpad=30)
ax1.tick_params(axis='y', width=2, labelsize=25, colors='#b60000')
ax.set_title('shopping market', weight='bold', fontsize=40)
```
